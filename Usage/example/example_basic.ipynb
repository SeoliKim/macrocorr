```python
import macrocorr as mcrr
```

```python
import pandas as pd
```

```python
apple_stock_dataset=pd.read_csv('AAPL.csv')
print (apple_stock_dataset.head())
```

```python
my_correlator= mcrr.Correlator(date=apple_stock_dataset['Date'].values, data_x=apple_stock_dataset['Close'].values)
```

```python
my_correlator.analyze_Correlation(category='commodity')
```

```python
my_correlator.get_Correlation(y_name="Net Financial Account", graph=True)
```
